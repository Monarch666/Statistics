# Quartile

Quartiles are three values that split sorted data into four parts, each with an equal number of observations. Quartiles are a type of quantile.

First quartile: Also known as Q1, or the lower quartile. This is the number halfway between the lowest number and the middle number.
Second quartile: Also known as Q2, or the median. This is the middle number halfway between the lowest number and the highest number.
Third quartile: Also known as Q3, or the upper quartile. This is the number halfway between the middle number and the highest number.

# Percentiles

A percentile is a statistical measure used to describe the relative position of a particular data point within a dataset or distribution. It helps to understand how a specific value compares to the rest of the values in the dataset. Percentiles are commonly used in various fields, including education, healthcare, and statistics.

Here are some common percentiles:

Median (50th Percentile): The median is the value that separates the lower 50% of the data from the upper 50%.
Quartiles: Quartiles divide the data into four equal parts, with the first quartile being the 25th percentile (Q1), the second quartile being the median (Q2), and the third quartile being the 75th percentile (Q3).
Percentiles: You can choose any percentile you want, such as the 10th, 90th, or any other value, to understand where a particular data point falls in the distribution.
Percentiles are especially useful when dealing with large datasets or when you want to understand the spread of data or identify outliers. For example, if a student scores in the 90th percentile on a standardized test, it means their score is higher than 90% of the students who took the test.

Example:

Suppose you want to estimate the percentage of defective items in a factory that produces 10,000 items. You decide to use simple random sampling to select a sample of 200 items and calculate the sample proportion of defects. 

Population Size (N): There are 10,000 items in the factory.

Sample Size (n): You want to select a sample of 200 items.

Sampling: Use a random number generator to select 200 unique items from the population.

Data Collection: Inspect the selected 200 items and count the number of defective items.

Sample Proportion (p̂)(Probability): Calculate the sample proportion of defects using the formula:


Calculations: Let's assume you found that 15 of the 200 items in the sample were defective.

p̂ = 15 / 200

p̂ = 0.075 (or 7.5%)

So, the sample proportion of defective items in the factory, based on the 200-item sample, is 7.5%. This estimate can be used to make inferences about the proportion of defects in the entire factory production.

```python
import numpy as np

# Define your data and parameters
data = np.array([85, 90, 88, 92, 78, 87, 89, 91, 82, 80])
mean = np.mean(data)
std_dev = np.std(data)

# Calculate the z-score for a specific data point
data_point = 85
z_score = (data_point - mean) / std_dev
print(mean)
print(std_dev)
print(f"Z-score for {data_point} is {z_score}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Define your data
data = np.array([85, 90, 88, 92, 78, 87, 89, 91, 82, 80])

# Calculate the z-score for the data
mean = np.mean(data)
std_dev = np.std(data)
z_scores = [(x - mean) / std_dev for x in data]

# Create a figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Plot a KDE plot for the original data in the first subplot
sns.kdeplot(data, shade=True, ax=axes[0])
axes[0].set_xlabel('Scores')
axes[0].set_ylabel('Density')
axes[0].set_title('KDE Plot of Original Data')

# Plot a KDE plot for the z-scores in the second subplot
sns.kdeplot(z_scores, shade=True, ax=axes[1])
axes[1].set_xlabel('Z-Scores')
axes[1].set_ylabel('Density')
axes[1].set_title('KDE Plot of Data After Applying Z-Scores')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Define your data
data = np.array([85, 90, 88, 92, 78, 87, 89, 91, 82, 80])

# Calculate the z-scores using scipy.stats.zscore
z_scores = stats.zscore(data)

# Create a figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Plot the original data in the first subplot
axes[0].hist(data, bins=10, edgecolor='k', alpha=0.7, color='skyblue', density=True)
axes[0].set_xlabel('Scores')
axes[0].set_ylabel('Probability Density')
axes[0].set_title('Data Before Applying Z-Scores')

# Plot the z-scores in the second subplot
axes[1].hist(z_scores, bins=10, edgecolor='k', alpha=0.7, color='lightcoral', density=True)
axes[1].set_xlabel('Z-Scores')
axes[1].set_ylabel('Probability Density')
axes[1].set_title('Data After Applying Z-Scores')

plt.tight_layout()
plt.show()
```

# Covariance

### Difference between correlation and covariance

```python
image_path = 'https://image1.slideserve.com/3122318/covariance-and-correlation-n.jpg'
html_code = f'<img src="{image_path}" alt="Image" style="width: 600px;">'
display(HTML(html_code))

```

```python
import numpy as np
import pandas as pd

# Sample data
hours_studied = [4, 7, 2, 8, 6, 5, 2, 8, 9, 3]
exam_scores = [85, 90, 70, 95, 80, 88, 75, 96, 98, 78]

# Calculate the covariance
covariance = np.cov(hours_studied, exam_scores)[0, 1]

# Calculate the correlation
correlation = np.corrcoef(hours_studied, exam_scores)[0, 1]

print(f'Covariance: {covariance}')
print(f'Correlation: {correlation}')
```

```python
import scipy.stats as stats

# Sample data for two groups (e.g., two samples)
group1 = [32, 40, 45, 50, 38]
group2 = [28, 35, 42, 48, 37]

# Perform an independent two-sample t-test
t_stat, p_value = stats.ttest_ind(group1, group2)

# Output the results
print("T-statistic:", t_stat)
print("P-value:", p_value)

# Interpret the results
alpha = 0.05  # significance level
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference between the two groups.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference between the two groups.")
```

# 2.Z-test

```python
import numpy as np
from scipy import stats

# Define your sample data and parameters
sample_data = np.array([82, 85, 88, 91, 94, 97])
population_mean = 90  # Known population mean
population_stddev = 5  # Known population standard deviation
sample_size = len(sample_data)

# Calculate the sample mean and standard error
sample_mean = np.mean(sample_data)
standard_error = population_stddev / np.sqrt(sample_size)

# Perform the Z-test
z_score = (sample_mean - population_mean) / standard_error
print(z_score)
# Set the significance level (alpha) for your test
alpha = 0.05

# Calculate the p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# Check if the p-value is less than alpha to determine significance
if p_value < alpha:
    print(f"The Z-test is statistically significant. p-value = {p_value:.4f}")
else:
    print(f"The Z-test is not statistically significant. p-value = {p_value:.4f}")
```

```python
from scipy import stats

# Given information
t_statistic = 3.47
df = 14

# Calculate the two-tailed p-value
p_value = 2 * (1 - stats.t.cdf(abs(t_statistic), df=df))

# Print the p-value
print("P-value:", p_value)
```

# 3.ANNOVA test

```python
import scipy.stats as stats

# Sample data for three or more groups
group1 = [10,9,8,7.5,8.5,9,10,8,8,9]
group2 = [8,9,10,8,8.5,7,9.5,9,7,10]
group3 = [9,8,7,10,9,8,7,10,9,8]

# Perform one-way ANOVA
f_statistic, p_value = stats.f_oneway(group1, group2, group3)

# Output the results
print("F-statistic:", f_statistic)
print("P-value:", p_value)

# Interpret the results
alpha = 0.05  # significance level
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference between the groups.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference between the groups.")
```
